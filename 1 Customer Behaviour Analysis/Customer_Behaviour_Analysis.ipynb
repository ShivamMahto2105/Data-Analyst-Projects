```python
"""There is null value in Review Rating column. We can replace by Median value but in the Customer Data it's not the best. Because rating can varies a lot 
depending on the category. Clothing Category can have different review rating than footwear category. Now, if we just take the overall median and fill it 
everywhere, we are basically forcing cothing items to get the same review ratings as footwear item. And that can introduce bias into our dataset. Instead,
what we will do is impute the missing values using the median review rating within each category. That's way if a clothing item is missing its review 
rating, it's get replaced with the median review rating of a clothing item, not the footwear item."""
```

```python
# This code fills missing values in the 'Review Rating' column using the median rating of each category
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
""" Convert all names into snake cast : Naming Convention in programming where spaces between words are replaced with an underscore(_) and all the 
letters are typically in lowercase. Make them more readable"""
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
# Check all updated columns 
df.columns
```

```python
# Create a new columns age_group 
labels  = ['Young Adult' , 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Create  column purchase_frequency_days

frquency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Anually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frquency_mapping)
```

```python
# Check the column of purchase_frequency_days
df[['purchase_frequency_days'  , 'frequency_of_purchases']].head(10)
```
